```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as sts
import numpy as np
```

```python
#Cell 300 by Nicole Langford

#read the csvs that need to be analyzed
all_countries_nl_df = pd.read_csv("Output/no_nans_df.csv")
all_countries_nl_df
```

```python
#cell 301 by Nicole Langford
#get name, total deaths, total crude deaths, year, sex into a database

new_death_data_df = all_countries_nl_df[["name", "Year", "Total Deaths", "Total Crude Death Rate", "Sex Label"]]
new_death_data_df
```

```python
#cell 302 by Nicole Langford

nz_deaths_df = new_death_data_df.loc[new_death_data_df["name"]== "New Zealand"]
nz_deaths_df
```

```python
#cell 303 by Nicole Langford
male_nz_deaths_df = nz_deaths_df.loc[nz_deaths_df["Sex Label"]== "Male"]
male_nz_deaths_df
```

```python
#cell 304 by Nicole Langford
female_nz_deaths_df = nz_deaths_df.loc[nz_deaths_df["Sex Label"]== "Female"]
female_nz_deaths_df
```

```python
#cell 305 by Nicole Langford
#creating new dataframe by gender
#create new variables for year and CDRs

NZ_Year = nz_deaths_df["Year"]
NZ_M_CDR = male_nz_deaths_df["Total Crude Death Rate"]
NZ_F_CDR = female_nz_deaths_df["Total Crude Death Rate"]


nz_crude_death_gender_dict = {"Year": NZ_Year,
                            "male crude death rate": NZ_M_CDR,
                            "female crude death rate": NZ_F_CDR
                           }
nz_crude_death_gender_df = pd.DataFrame(nz_crude_death_gender_dict)
nz_crude_death_gender_df
```

```python
#cell 306 by Nicole Langford
nz_total_crude_deaths_df = nz_crude_death_gender_df.groupby("Year").sum().reset_index()
nz_total_crude_deaths_df
```

```python
#cell 307 by Nicole Langford
x_axis = nz_total_crude_deaths_df["Year"]
y_axis_m = nz_total_crude_deaths_df["male crude death rate"]
y_axis_f = nz_total_crude_deaths_df["female crude death rate"]

nz_male_cdr, = plt.plot(x_axis, y_axis_m, color= "green", label= "Male Death Rate")
nz_female_cdr, = plt.plot(x_axis, y_axis_f, color= "purple", label= "Female Death Rate")
plt.legend(handles= [nz_male_cdr, nz_female_cdr], loc="best")
plt.title("New Zealand Yearly Crude Death Rates By Gender 2000-2015")
plt.xlabel("Year")
plt.ylabel("Deaths Per 1000 People")
plt.grid()

#save image
plt.savefig("Output/New Zealand Yearly Crude Death Rates By Gender 200-2015")
plt.show()

```

```python
#cell 308 by Nicole Langford
#create graph to drill down on years 2004-2006
nz_2004_data_df = nz_total_crude_deaths_df.loc[(nz_total_crude_deaths_df["Year"] == 2004) | (nz_total_crude_deaths_df["Year"] == 2006) | (nz_total_crude_deaths_df["Year"] == 2005)]


x_axis = nz_2004_data_df["Year"]
y_axis_m = nz_2004_data_df["male crude death rate"]
y_axis_f = nz_2004_data_df["female crude death rate"]

nz_male_cdr, = plt.plot(x_axis, y_axis_m, color= "green", label= "Male Death Rate")
nz_female_cdr, = plt.plot(x_axis, y_axis_f, color= "purple", label= "Female Death Rate")
plt.legend(handles= [nz_male_cdr, nz_female_cdr], loc="best")
plt.title("New Zealand Yearly Crude Death Rates By Gender, 2004-2006")
plt.xlabel("Year")
plt.ylabel("Deaths Per 1000 People")
plt.grid()

#save image
plt.savefig("Output/New Zealand Yearly Crude Death Rates By Gender, 2004-2006")
plt.show()
```

```python
#cell 309 by Nicole Langford
#creating new dataframe by gender and elderly pop
#create new variables for year and CDRs

all_elderly_df = all_countries_nl_df[["name", "Year", "Total Deaths", "Total Crude Death Rate", "Sex Label", "Older Elderly Age 80 and Above"]]
all_elderly_df

nz_eldery_deaths_df = all_elderly_df.loc[all_elderly_df["name"]== "New Zealand"]
nz_eldery_deaths_df

male_elderly_nz_df = nz_eldery_deaths_df.loc[nz_eldery_deaths_df["Sex Label"]== "Male"]
male_elderly_nz_df

female_elderly_nz_df = nz_eldery_deaths_df.loc[nz_eldery_deaths_df["Sex Label"]== "Female"]
female_elderly_nz_df



NZ_Year = nz_deaths_df["Year"]
nz_male_elder_deaths = male_elderly_nz_df["Older Elderly Age 80 and Above"]
nz_female_elder_deaths = female_elderly_nz_df["Older Elderly Age 80 and Above"]


nz_elder_death_gender_dict = {"Year": NZ_Year,
                            "male elderly death rate": nz_male_elder_deaths,
                            "female elderly death rate": nz_female_elder_deaths
                           }
nz_elder_death_gender_df = pd.DataFrame(nz_elder_death_gender_dict)
nz_elder_death_gender_df
```

```python
# cell 310 
#chart elderly deaths by gender
nz_elder_death_gender_df = nz_elder_death_gender_df.groupby("Year").sum().reset_index()
nz_elder_death_gender_df
```

```python
#cell 311 by Nicole Langford
x_axis = nz_elder_death_gender_df["Year"]
y_axis_m = nz_elder_death_gender_df["male elderly death rate"]
y_axis_f = nz_elder_death_gender_df["female elderly death rate"]

NZ_Year = nz_deaths_df["Year"]
nz_male_elder_deaths = male_elderly_nz_df["Older Elderly Age 80 and Above"]
nz_female_elder_deaths = female_elderly_nz_df["Older Elderly Age 80 and Above"]

nz_male_elder_deaths, = plt.plot(x_axis, y_axis_m, color= "green", label= "Male Elderly Death Rate")
nz_female_elder_deaths, = plt.plot(x_axis, y_axis_f, color= "purple", label= "Female Elderly Death Rate")
plt.legend(handles= [nz_male_elder_deaths, nz_female_elder_deaths], loc="best")
plt.title("New Zealand Yearly Elderly Death Rates By Gender 2000-2015")
plt.xlabel("Year")
plt.ylabel("Deaths Per 1000 People")
plt.grid()

#save image
plt.savefig("Output/New Zealand Yearly Elderly Death Rates By Gender 2000-2015")
plt.show()
```
